```python
# prime number check

n = int(input())

if n < 2:
  print("not prime")

else:
  for i in range(2,n):
    if n % i == 0:
      print("not prime")
      break
  else:
    print("prime")
```

```python
#Count Primes from 1 to N
n = int(input())

count = 0

for i in range (2,n+1):
      for j in range(2,i):
        if i % j == 0:
          break
      else:
        count += 1
print(count)
```

```python
# Palindrome Number
num = int(input())
temp = num
rev = 0


while temp != 0:
  last_digit = temp % 10
  rev = rev * 10+ last_digit
  temp = temp//10


if rev == num :
  print("palindrome")
else:
  print("not palindrome")
```

```python
1# Reverse Number
n = int(input())
rev = 0
while n != 0:
  last_digit = n % 10
  rev = rev * 10 + last_digit
  n = n//10
print(rev)
```

```python
#Count Digits
n = int(input())
count = 0
while n != 0:
  last_digit = n % 10
  count += 1
  n = n//10
print(count)
```

```python
# sum of digits
n = int(input())
sum = 0

while n != 0:
  last_digit = n%10
  sum = sum + last_digit
  n = n//10
print(sum)
```

```python
# factorial
n = int(input())
fact = 1
for i in range (1,n+1):
  fact = fact * i
print(fact)
```

```python
# fibonacci number
n = int(input())
a,b = 0,1

for i in range(n):
  print(a,end=" ")
  a,b = b,a+b
```

```python
# Armstrong number
num  = int(input())
temp = num
digits = len(str(num))
sum = 0

while temp !=0:
  d = temp % 10
  sum = sum +d**digits
  temp = temp//10
print("amrstrong "if sum == num else "not armstrong")
```

```python
# find max and min in the list
lst = [10,30,40]
mx = lst[0]
mn = lst[0]

for x in lst:
  if x > mx : mx = x
  if x < mn : mn = x
print(mx,mn)
```

```python
def even_or_odd (num):
  if num%2==0:
    return "the number {} is even ".format(num)
  else:
    return "the number {} is odd".format(num)
```

```python
even_or_odd(4)
```

```python
map(even_or_odd,lst)
```
